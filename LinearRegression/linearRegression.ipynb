### Linear Regression Demo

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

#### training model

In sample space $X = \{x_1,x_2,\cdots,x_m\}$,each $x_i$ has a label $y_i$, formed a label set $Y$. In linear regression we try to find a linear model which is
$$\hat{y} = \theta_0 + \theta_1 x$$
to make it.

```python
X = np.array([2,3,5,6,8,10,11,13]).reshape(-1,1)
y = np.array([3,5,7,9,11,13,15,17])

model = LinearRegression()

model.fit(X,y)
```

#### draw figure

```python
plt.scatter(X,y,color='blue')
plt.plot(X,model.predict(X),color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
plt.show()
print("y =", model.intercept_, "+", model.coef_[0], "* x")
```
